```python
# Request the page 
import urllib.request
url = "https://www.denvergov.org/content/denvergov/en/denver-development-services/help-me-find-/building-permits.html"
html_content = urllib.request.urlopen(url)
# Fetching html content
soup = BeautifulSoup(html_content)
pdf_links=[]
xlsx_links=[]
other_links=[]
# Fetching only href attributes 
for div_tag in soup.find_all('div',class_="panel-body"):
    for a_tag in div_tag('a'):
        link = a_tag.get('href')
        if link.startswith('https:') and link.endswith('.pdf'):
            pdf_links.append(link)
        elif link.endswith('.pdf'): 
            link = "https://www.denvergov.org"+link
            pdf_links.append(link)

        elif link.startswith('https:') and link.endswith('.xls'):
            xlsx_links.append(link)
        elif link.endswith('.xls'): 
            link = "https://www.denvergov.org"+link
            xlsx_links.append(link)
        else:
            other_links.append(link)
```

```python
import os
import openpyxl
import pandas as pd
os.chdir()
from openpyxl import load_workbook
wb = load_workbook('April-2016.xlsx')
ws =wb.active
max_row=ws.max_row
max_row

# Fetching rows where Stat Code is mentioned
stat_list_values = {}
for rows in ws.iter_rows("A"):
    for cell in rows:
        if "Stat Code" in str(cell.value):
            stat_list_values[cell.row] = cell.value     
#stat_list_values

# Get the values in column A as a list
colA_values = []
for rows in ws.iter_rows("A"):
    for cell in rows:
        if "Stat Code" in str(cell.value):
            cell.value = " "
        colA_values.append(cell.value)
#colA_values  

# Get the Permit Type values
permit_type_list_values = {}
for rows in ws.iter_rows("B"):
    for cell in rows:
        if "Permit Type" in str(cell.value):
            permit_type_list_values[cell.row] = cell.value 
#permit_type_list_values

# Getting the values of columns in a dictionary with column names as keys and column values as values.
d = {}
for cols in ["D","E","F","G","H","I","J","K","L","M","N","O","P","Q","R","S","T","U"]:
    nonevals = []
    for rows in ws.iter_rows(cols):
        for cell in rows:
                nonevals.append(cell.value)
    d[cols] = nonevals
    
# Assigning the column A values to key A
d["A"] = colA_values

# Converting the 
df = pd.DataFrame(d)
df.head()

# Creating and filling Stat Code column
df['Stat Code'] = " "
keys = list(stat_list_values.keys())
df.iloc[0:keys[0],-1] = stat_list_values[keys[0]]
for i in range(len(keys)-1):
      df.iloc[keys[i]:keys[i+1],-1] = stat_list_values[keys[i]]
        
# Creating and filling Permit Type column
df['Permit Type'] = " "
keys = list(permit_type_list_values.keys())
df.iloc[0:keys[0],-1] = permit_type_list_values[keys[0]]
for i in range(len(keys)-1):
      df.iloc[keys[i]:keys[i+1],-1] = permit_type_list_values[keys[i]]
        
# Replacing NAs with blanks and then removing rows where Permit # is missing
df.fillna('',inplace=True)
df1 = df[df["D"] != ""]

# COlumn names from the excel file
excelcols = list(df1.columns)

#zipping the two column names in a dictionary
cols = {}
for col1,col2 in zip(excelcols,colnames):
    cols[col1] = col2
#cols

# columns to drop from the dataframe as they don't have a actual column name
to_drop = []
for key in cols:
    if cols[key] =="":
        to_drop.append(key)
to_drop

# Drop those columns from dataframe
df1.drop(to_drop,axis=1,inplace=True)

# Deleting those keys from the rename dictionary
for key in list(cols.keys()):
    if cols[key] == "":
        del cols[key]
cols['Stat Code'] = "Stat Code"
cols['Permit Type'] = "Permit Type"

# Rename the columns in the dataframe
df1.rename(columns = cols, inplace = True)

# Filter out the first row containing the column names
df_final = df1.iloc[1:,]

# Write the formatted data frame to a csv
df_final.to_csv('April-2016_processed.csv',index = False)

df_final_db.to_sql('denver_permits_final', engine)
```
